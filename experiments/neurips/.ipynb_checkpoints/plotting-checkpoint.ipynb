```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results/exact_vs_approx.csv')
df = df[df.iters ]
exact = df[df.oracle == 'exact'].groupby('measurements')
approx = df[df.oracle == 'approx'].groupby('measurements')

exact.error_in.mean().plot(linewidth=3, label='Exact')
approx.error_in.mean().plot(linewidth=3, label='Approx')
plt.xlabel('Number of Measurements', fontsize='x-large')
plt.ylabel('Error on Measured Cliques', fontsize='x-large')
plt.legend(fontsize='x-large')
```

```python
exact.error_out.median().plot(linewidth=3, label='Exact')
approx.error_out.median().plot(linewidth=3, label='Approx')
plt.xlabel('Number of Measurements', fontsize='x-large')
plt.ylabel('Error on Unmeasured Cliques', fontsize='x-large')
plt.legend(fontsize='x-large')
```
