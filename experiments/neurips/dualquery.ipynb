```python
from dpsynth.mechanisms import DualQuery, FEM
from mbi import Dataset, CliqueVector, FactoredInference
import itertools
import numpy as np
%load_ext autoreload 
%autoreload 2
```

```python
prefix = '/home/rmckenna/Repos/hd-datasets/clean/'

def load(name):
    return Dataset.load(prefix + name + '.csv', prefix + name + '-domain.json')

data = load('adult')
```

```python
cliques = list(itertools.combinations(data.domain, 2))[:5]
W = [(cl, None) for cl in cliques]

mech = DualQuery(100, 0.01)
synth = mech.run(data, W)
```

```python
baseline = Dataset.synthetic(data.domain, data.records)
for cl in cliques:
    x = data.project(cl).datavector()
    y = synth.project(cl).datavector()
    z = baseline.project(cl).datavector()
    x /= x.sum()
    y /= y.sum()
    z /= z.sum()
    print(cl, np.linalg.norm(x-y,1), np.linalg.norm(x-z,1))
```
