```python
from mbi import Domain, RegionGraph
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edge('ABC','A')
G.add_edge('ABC','BC')
G.add_edge('AD','A')
G.add_edge('AD','D')
G.add_edge('BCD','BC')
G.add_edge('BCD','D')
G.add_edge('DE','D')

pos = { 'ABC' : (0,1), 'AD' : (1,1), 'BCD': (2,1), 'DE' : (3,1), 
       'A' : (0.5,0), 'BC' : (1.5,0), 'D':(2.5,0) }

node_color = ['#33AAFF','#3CB371']
node_color = [node_color[pos[cl][1]] for cl in G.nodes]

fig = plt.figure(figsize=(5,2.5))
nx.draw(G, pos=pos, 
        node_size=1600, 
        node_color=node_color, #'#3CB371', 
        with_labels=True,
        width=2.5,
        font_weight='bold')

plt.ylim([-0.2, 1.2])
plt.savefig('ex_region.pdf')
```

```python
G = nx.DiGraph()
G.add_edge('ABC','A')
G.add_edge('ABC','B')
G.add_edge('ABC','C')
G.add_edge('AD','A')
G.add_edge('AD','D')
G.add_edge('BCD','B')
G.add_edge('BCD','C')
G.add_edge('BCD','D')
G.add_edge('DE','D')

pos = { 'ABC' : (0,1), 'AD' : (1,1), 'BCD': (2,1), 'DE' : (3,1), 
       'A' : (0,0), 'B' : (1,0), 'C':(2,0), 'D':(3,0) }

node_color = ['#33AAFF','#3CB371']
node_color = [node_color[pos[cl][1]] for cl in G.nodes]

fig = plt.figure(figsize=(5,2.5))
nx.draw(G, pos=pos, 
        node_size=1500, 
        node_color=node_color,
        with_labels=True,
        width=2.5,
        font_weight='bold')

plt.ylim([-0.2, 1.2])
plt.savefig('ex_factor.pdf')
```

```python
G = nx.DiGraph()
G.add_edge('ABCD','D')
G.add_edge('DE','D')

pos = { 'ABCD' : (0,1), 'DE' : (1,1), 'D':(0.5,0) }
node_color = ['#33AAFF','#3CB371']
node_color = [node_color[pos[cl][1]] for cl in G.nodes]

fig = plt.figure(figsize=(2.5,2.5))
nx.draw(G, pos=pos, 
        node_size=1500, 
        node_color=node_color,
        with_labels=True,
        width=2.5,
        font_weight='bold')
plt.ylim([-0.2, 1.2])
plt.xlim([-0.2, 1.2])
plt.savefig('ex_clique.pdf')
```
