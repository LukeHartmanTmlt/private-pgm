```python
import numpy as np
```

```python
tAB = np.random.rand(2,3,1)*3
tAC = np.random.rand(2,1,4)*3
tBC = np.random.rand(1,3,4)*3

P = np.exp(tAB + tAC + tBC); P /= P.sum()

mAB = P.sum(axis=2, keepdims=True)
mAC = P.sum(axis=1, keepdims=True)
mBC = P.sum(axis=0, keepdims=True)

mA = P.sum(axis=(1,2), keepdims=True)
mB = P.sum(axis=(0,2), keepdims=True)
mC = P.sum(axis=(0,1), keepdims=True)

Q = np.ones(P.shape); Q /= Q.sum()

for _ in range(15):
    print(np.linalg.norm(Q-P))
    
    Q = Q * mAB / Q.sum(axis=2, keepdims=True)
    Q /= Q.sum()
    Q = Q * mAC / Q.sum(axis=1, keepdims=True)
    Q /= Q.sum()
    Q = Q * mBC / Q.sum(axis=0, keepdims=True)
    Q /= Q.sum()
```

```python
#mAB = np.random.rand(2,3,1); mAB /= mAB.sum()
#mAC = np.random.rand(2,1,4); mAC /= mAC.sum()
#mBC = np.random.rand(1,3,4); mBC /= mBC.sum()

Q = np.ones((2,3,4)); Q /= Q.sum()
alpha = 1.0

def err(Q):
    nAB = Q.sum(axis=2, keepdims=True)
    nAC = Q.sum(axis=1, keepdims=True)
    nBC = Q.sum(axis=0, keepdims=True)
    return np.sum((nAB-mAB)**2)+np.sum((nAC-mAC)**2)+np.sum((nBC-mBC)**2)

for _ in range(100):
    print(err(Q))
    nAB = Q.sum(axis=2, keepdims=True)
    nAC = Q.sum(axis=1, keepdims=True)
    nBC = Q.sum(axis=0, keepdims=True)
    diff = (nAB - mAB) + (nAC - mAC) + (nBC - mBC)
    Q = Q * np.exp(-alpha*diff)
    Q /= Q.sum()
```
